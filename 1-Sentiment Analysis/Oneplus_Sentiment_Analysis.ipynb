```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(df['Tweet_Length'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Tweet Lengths")
plt.xlabel("Tweet Length")
plt.ylabel("Frequency")
plt.show()
```

```python
from collections import Counter
import re

# Combine all tweets into one string
all_text = ' '.join(df['Tweet'].astype(str))

# Clean text: remove URLs, hashtags, mentions, punctuations
words = re.findall(r'\b\w+\b', all_text.lower())

# Count top 20 most common words
common_words = Counter(words).most_common(20)

# Convert to DataFrame
words_df = pd.DataFrame(common_words, columns=['Word', 'Frequency'])

# Plot
plt.figure(figsize=(12,6))
sns.barplot(data=words_df, x='Frequency', y='Word', palette='viridis')
plt.title("Top 20 Most Common Words in Tweets")
plt.xlabel("Frequency")
plt.ylabel("Words")
plt.show()
```

```python
# Convert 'Time' column to datetime
df['Time'] = pd.to_datetime(df['Time'])

# Group tweets per hour/day
tweet_times = df.set_index('Time').resample('H').count()

# Plot
plt.figure(figsize=(12,6))
tweet_times['Tweet'].plot()
plt.title("Tweet Volume Over Time")
plt.xlabel("Time")
plt.ylabel("Number of Tweets")
plt.grid(True)
plt.show()
```

## 5. Data Cleaning

```python
import re

def clean_tweet(text):
    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    
    # Remove mentions and hashtags
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#\w+', '', text)
    
    # Remove emojis and non-ASCII characters
    text = text.encode('ascii', 'ignore').decode('ascii')
    
    # Remove newline characters and extra whitespace
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

# Apply to your dataset
df['Clean_Tweet'] = df['Tweet'].apply(clean_tweet)

# Preview the cleaned text
df[['Tweet', 'Clean_Tweet']].head()
```

## 7.Sentiment Prediction

```python
!pip install textblob
```

```python
from textblob import TextBlob

def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return "Positive"
    elif analysis.sentiment.polarity < 0:
        return "Negative"
    else:
        return "Neutral"

# Apply to Cleaned Tweets
df['Sentiment'] = df['Clean_Tweet'].apply(get_sentiment)

# Show results
df[['Clean_Tweet', 'Sentiment']].head()
```

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.countplot(x='Sentiment', data=df, palette='coolwarm')
plt.title("Sentiment Distribution in OnePlus Tweets")
plt.xlabel("Sentiment")
plt.ylabel("Number of Tweets")
plt.show()
```
