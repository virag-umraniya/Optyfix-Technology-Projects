# Opportunity Radar (Radar Chart)

```python
import numpy as np

labels = np.array(["Customization", "Certification", "Regional Reach", "AI Integration", "Mobile Learning"])
values = [8, 7, 9, 6, 9]

angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
values += values[:1]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angles, values, 'o-', linewidth=2)
ax.fill(angles, values, alpha=0.25)
ax.set_thetagrids(np.degrees(angles[:-1]), labels)
ax.set_title("OPTIFYX Opportunity Radar")
plt.show()
```
